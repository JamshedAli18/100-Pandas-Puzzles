```python
import matplotlib.pyplot as plt
import pandas as pd
# Create initial DataFrame for plotting
data = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 35]}
df = pd.DataFrame(data)
df
```

```python
# Puzzle 1: Plot a line graph.
plt.figure(figsize=(8, 4))
plt.plot(df['Name'], df['Age'], marker='o')
plt.title('Age by Name')
plt.xlabel('Name')
plt.ylabel('Age')
plt.show()



```

```python
# Puzzle 2: Plot a bar graph.
plt.figure(figsize=(8, 4))
plt.bar(df['Name'], df['Age'], color='skyblue')
plt.title('Age by Name')
plt.xlabel('Name')
plt.ylabel('Age')
plt.show()
```

```python
# Puzzle 3: Plot a histogram.
plt.figure(figsize=(8, 4))
plt.hist(df['Age'], bins=5, color='green')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python

# Puzzle 4: Plot a scatter plot.
data = {'Height': [5.1, 6.2, 5.9], 'Weight': [60, 75, 72]}
df = pd.DataFrame(data)
plt.figure(figsize=(8, 4))
plt.scatter(df['Height'], df['Weight'], color='red')
plt.title('Height vs Weight')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()

```

```python
# Puzzle 5: Plot multiple lines on the same graph.
data = {'Month': ['Jan', 'Feb', 'Mar'], 'Temperature_A': [30, 35, 40], 'Temperature_B': [25, 30, 35]}
df = pd.DataFrame(data)
plt.figure(figsize=(8, 4))
plt.plot(df['Month'], df['Temperature_A'], marker='o', label='City A')
plt.plot(df['Month'], df['Temperature_B'], marker='x', label='City B')
plt.title('Monthly Temperature')
plt.xlabel('Month')
plt.ylabel('Temperature')
plt.legend()
plt.show()

```

```python
# Puzzle 6: Plot a pie chart.
data = {'Category': ['A', 'B', 'C'], 'Values': [30, 40, 30]}
df = pd.DataFrame(data)
plt.figure(figsize=(8, 4))
plt.pie(df['Values'], labels=df['Category'], autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99'])
plt.title('Category Distribution')
plt.show()

```

```python
# Puzzle 7: Plot a box plot.
data = {'Category': ['A', 'A', 'B', 'B', 'C', 'C'], 'Values': [30, 35, 40, 45, 50, 55]}
df = pd.DataFrame(data)
plt.figure(figsize=(8, 4))
df.boxplot(by='Category')
plt.title('Box Plot by Category')
plt.suptitle('')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()

```

```python
# Puzzle 8: Plot a heatmap.
import numpy as np
data = np.random.rand(10, 10)
plt.figure(figsize=(8, 6))
plt.imshow(data, cmap='hot', interpolation='nearest')
plt.colorbar()
plt.title('Heatmap')
plt.show()

```
