```python
import pandas  as pd
```

```python
# Creating a sample DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 35], 'Birthdate': ['1995-01-01', '1990-05-15', '1985-09-30']}
df = pd.DataFrame(data)
```

```python
# Puzzle 1: Select rows where a column's value is in a list.
df_selected = df[df['Name'].isin(['Alice', 'Charlie'])]
df_selected
```

```python
# Puzzle 2: Apply a function to each row.
df['Age_doubled'] = df['Age'].apply(lambda x: x * 2)
df
```

```python
# Puzzle 3: Merge two DataFrames on a common column.
df1 = pd.DataFrame({'Name': ['Alice', 'Bob'], 'Age': [25, 30]})
df2 = pd.DataFrame({'Name': ['Alice', 'Bob'], 'City': ['New York', 'Los Angeles']})
df_merged = pd.merge(df1, df2, on='Name')
df_merged

```

```python
# Puzzle 4: Pivot a DataFrame.
df['City'] = ['New York', 'Los Angeles', 'Chicago']
df_pivoted = df.pivot(index='Name', columns='Age', values='City')
df_pivoted
```

```python
# Puzzle 5: Unpivot a DataFrame.
df_unpivoted = df.melt(id_vars=['Name'], value_vars=['Age', 'City'])
df_unpivoted
```

```python
# Puzzle 6: Replace missing values in a DataFrame.
df_with_na = df.copy()
df_with_na.loc[1, 'Age'] = None
df_filled = df_with_na.fillna(0)
df_filled
```

```python
# Puzzle 7: Create a new column based on a condition.
df['Senior'] = df['Age'] > 30
df
```

```python

# Puzzle 8: Calculate the cumulative sum of a column.
df['Cumulative_Age'] = df['Age'].cumsum()
df
```

```python

# Puzzle 9: Use applymap to modify all elements of a DataFrame.
df_multiplied = df[['Age']].applymap(lambda x: x * 2)
print(df_multiplied)
```
